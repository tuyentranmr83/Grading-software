```python
#Task 2
def analyze_grades(file_name):
    try:
        with open(f'data-files/Data Files/{file_name}.txt', 'r') as file:
            #If exists print notitfication
            print(f'Successfully opened {file_name}')
            lines = file.readlines()

            # Initialize counters
            valid_lines = 0
            invalid_lines = 0
            # Print the analysis report
            print("\n**** ANALYZING ****\n")

            # Loop through each line
            for line in lines:
                # Split the line by comma
                data = line.strip().split(',')
                # Check if the line has exactly 26 values and starts with 'N'
                if len(data) == 26 and data[0].startswith('N') and data[0][1:].isdigit() and len(data[0]) == 9:
                    #Add value to valid_lines
                    valid_lines += 1
                else:
                    if len(data) != 26:
                        print(f"Invalid line of data: does not contain exactly 26 values:\n\n {line}")
                    elif not data[0].startswith('N') or not data[0][1:].isdigit():                                            
                        print(f"Invalid line of data: N# is invalid:\n\n {line}")
                    elif len(data[0]) != 9:
                        print(f"Invalid line of data: ID student is not 9 characters:\n\n {line}")                       
                    #Add value to invalid_lines
                    invalid_lines += 1
            
            if invalid_lines == 0:
                print("No errors found!\n")
                
            print("**** REPORT ****\n")
            print(f"Total valid lines of data: {valid_lines}\n")
            print(f"Total invalid lines of data: {invalid_lines}\n")

    except FileNotFoundError:
        print(f"Error: File '{file_name}' not found.")
    except Exception as e:
        print(f'Error: {e}')

# Get user input for the file name
file_name = input("Enter a class to grade (e.g., class1 for class1.txt): ")

# Call the function to analyze grades
analyze_grades(file_name)
```

```python
#Task 3

def analyze_grades_exam(file_name):
    try:
        with open(f'data-files/Data Files/{file_name}.txt', 'r') as file:
            #If exists print notitfication
            print(f'Successfully opened {file_name}')
            lines = file.readlines()

            # Initialize counters
            valid_lines = 0
            invalid_lines = 0
            high_scores = 0
            scores = []
            
            # Initialize dictionaries to track skipped and incorrect questions
            skipped_questions = {}
            incorrect_questions = {}

            # Answer key
            answer_key = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D"
            answers_key = answer_key.split(',')
            
            # Print report
            print(f"\n**** ANALYZING ****\n")

            # Loop through each line
            for line in lines:
                # Split the line by comma
                data = line.strip().split(',')

                # Check if the line has exactly 26 values and starts with 'N'
                if len(data) == 26 and data[0].startswith('N') and data[0][1:].isdigit() and len(data[0]) == 9:
                    
                    #Add value to valid_lines
                    valid_lines += 1
                    
                    # Grade the exam
                    student_answers = data[1:]
                    score = 0
                    for i in range(len(answers_key)):
                        if student_answers[i] == answers_key[i]:
                            # correct answers + 4 score
                            score += 4
                        elif student_answers[i] == '':
                            # Skip question + 0 score
                            skipped_questions[i+1] = skipped_questions.get(i+1, 0) + 1
                        else:
                            # Incorrect answer - 1 score
                            incorrect_questions[i+1] = incorrect_questions.get(i+1, 0) + 1
                            score -= 1

                    # Update statistics
                    scores.append(score)
                    if score > 80:
                        high_scores += 1
                else:
                    if len(data) != 26:
                        print(f"Invalid line of data: does not contain exactly 26 values:\n\n {line}")
                    elif not data[0].startswith('N') or not data[0][1:].isdigit():                                            
                        print(f"Invalid line of data: N# is invalid:\n\n {line}")
                    elif len(data[0]) != 9:
                        print(f"Invalid line of data: ID student is not 9 characters:\n\n {line}")
                        
                    #Add value to invalid_lines
                    invalid_lines += 1

            # Calculate statistics            
            mean_score = round(sum(scores) / len(scores), 2)
            highest_score = max(scores)
            lowest_score = min(scores)
            score_range = highest_score - lowest_score
                        
            # Calculate median_scores
            sorted_scores = sorted(scores)
            n = len(sorted_scores)
            if n % 2 == 1:
                median_score = sorted_scores[n // 2]
            else:
                median_score = (sorted_scores[(n // 2) - 1] + sorted_scores[(n // 2)]) / 2
            
            median_scores = round(median_score, 2)

            # Find questions with most skips
            max_value = max(skipped_questions.values())
            skipped_most = [(key, value) for key, value in skipped_questions.items() if value == max_value]            

            # Find questions with most incorrect answers
            max_values = max(incorrect_questions.values())
            incorrect_questions_most = [(key, value) for key, value in incorrect_questions.items() if value == max_values]           
            
            if invalid_lines == 0:                
                print("No errors found!\n")
                
            print("**** REPORT ****\n")
            print(f"Total valid lines of data: {valid_lines}\n")
            print(f"Total invalid lines of data: {invalid_lines}\n")
            
            print(f"Total student of high scores: {high_scores}\n")
            print(f"Mean (average) score: {mean_score}\n")
            print(f"Highest score: {highest_score}\n")
            print(f"Lowest score: {lowest_score}\n")
            print(f"Range of scores: {score_range}\n")
            print(f"Median score: {median_scores}\n")
            print(f"Question that most people skip:")
            for skip_most in skipped_most:
                print(f"{skip_most[0]} - {skip_most[1]} - {round(skip_most[1]/valid_lines, 2)}", end = ', ')
            print(f"\n\nQuestion that most people answer incorrectly:")      
            for incorrect_question in incorrect_questions_most:
                print(f"{incorrect_question[0]} - {incorrect_question[1]} - {round(incorrect_question[1]/valid_lines, 2)}", end = ', ') 
           
    except FileNotFoundError:
        print(f"Error: Class file '{file_name}.txt' not found.")
    except Exception as e:
        print(f'Error: {e}')
        
# Get user input for the file name
file_name = input("Enter a class to grade (e.g., class1 for class1.txt): ")

# Call the function to analyze class

analyze_grades_exam(file_name)
```
